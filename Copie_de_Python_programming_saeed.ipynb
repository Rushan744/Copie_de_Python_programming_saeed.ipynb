# Jeu de hazard
Créer une fonction de jeu de hazard.
* Le programme choisit aléatoirement un chiffre entre 1 et 10 inclus.
* Le joueur choisit aussi un chiffre entre 1 et 10 inclus.
* Si le joueur a gagné,
  * le message "Félicitations, vous êtes l'heureux gagnant du jeu concours !" est affiché
  * et la fonction renvoie la valeur de gain "500 €".
* Si le joueur a perdu,
  * le message "Dommage, ce sera pour une prochaine fois." est affiché
  * et la fonction renvoie la valeur de gain "0 €".
* Dans tous les cas, la fonction affichera également le message "Merci de votre participation.".
* A votre avis quelle est la probabilité que le joueur remporte les 500€ en jouant une seule fois ?

```python
import random

def hazard_game():
    # Computer randomly chooses a digit between 1 and 10 inclusive
    computer_choice = random.randint(1, 10)

    # Player chooses a number between 1 and 10 inclusive
    player_choice = int(input("Enter your guess (a number between 1 and 10): "))

    # Check if the player has won or lost
    if player_choice == computer_choice:
        print("Congratulations, you're the happy winner of the contest game!")
        print("Thank you for your participation.")
        return 500  # Player wins $500
    else:
        print("Too bad, it will be for next time.")
        print("Thank you for your participation.")
        return 0  # Player loses, no gain

# Call the hazard_game function
gain = hazard_game()

# Print the probability of winning
probability_of_winning = 1 / 10  # There are 10 possible numbers (1 to 10)
print(f"The probability of winning $500 is: {probability_of_winning * 100}%")
```

# Compte à rebours

Créer un programme qui effectue un compte à rebours :
* En commençant par la valeur renseignée par l'utilisateur.
* En terminant à 0.
* Qui affiche le message "Il reste x opérations", x étant le nombre d'opérations restantes.
* Le message s'affiche uniquement lorsque x est un multiple de 5. Sinon aucun message n'est affiché.

```python
def countdown(start_value):
    """
    Perform a countdown starting from the given value and display messages
    when the remaining operations are multiples of 5.

    Parameters:
    - start_value (int): The starting value for the countdown.
    """
    # Validate that the input is a positive integer
    if not isinstance(start_value, int) or start_value <= 0:
        print("Please enter a valid positive integer.")
        return

    # Perform the countdown
    for i in range(start_value, -1, -1):
        # Display a message when the remaining operations are multiples of 5
        if i % 5 == 0:
            print(f"There are {i} operations remaining.")

    print("Countdown complete!")

# Get the starting value from the user
try:
    # Prompt the user to enter the starting value
    start_value = int(input("Enter the starting value for the countdown: "))
    # Call the countdown function with the entered value
    countdown(start_value)
except ValueError:
    # Handle the case where the user enters something that can't be converted to an integer
    print("Please enter a valid integer.")
```

#Distance 3D

* Créez une fonction calc_distance_3D() qui calcule la distance euclidienne en trois dimensions entre deux atomes.  
* Testez votre fonction sur les 2 points A(0,0,0) et B(1,1,1). Trouvez-vous bien √3 ?  
* On rappelle que la distance euclidienne d entre deux points A et B de coordonnées cartésiennes respectives (xA, yA,zA) et (xB, yB,zB) se calcule comme suit :  

formule_3D.png

```python
import math

def calc_distance_3D(xA, yA, zA, xB, yB, zB):
    """
    Calcule la distance euclidienne en trois dimensions entre deux atomes.

    Parameters:
    - xA, yA, zA (float): Coordonnées du premier atome (point A).
    - xB, yB, zB (float): Coordonnées du deuxième atome (point B).

    Returns:
    - float: La distance euclidienne entre les deux atomes.
    """
    distance = math.sqrt((xB - xA)**2 + (yB - yA)**2 + (zB - zA)**2)
    return distance

# Coordonnées des points A et B
xA, yA, zA = 0, 0, 0
xB, yB, zB = 1, 1, 1

# Calcul de la distance entre les points A et B
distance_AB = calc_distance_3D(xA, yA, zA, xB, yB, zB)

# Affichage du résultat
print(f"La distance euclidienne entre A et B est : {distance_AB}")

# Vérification si c'est bien √3
if math.isclose(distance_AB, math.sqrt(3)):
    print("La distance calculée est bien égale à √3.")
else:
    print("La distance calculée n'est pas égale à √3.")
```

# Distance de Hamming

La distance de Hamming mesure la différence entre deux séquences de même taille en comptant le nombre de positions
qui, pour chaque séquence, ne correspondent pas au même acide aminé.
* Créer la fonction `dist_hamming()` qui prend en argument deux chaînes de caractères et qui renvoie la distance de
Hamming (sous la forme d’un entier) entre ces deux chaînes de caractères.
* Calculez la distance de Hamming entre les séquences
`AGWPSGGASAGLAIL` et `IGWPSAGASAGLWIL`
* puis entre les séquences
`ATTCATACGTTACGATT` et `ATACTTACGTAACCATT`.

Exemple : la distance de Hamming entre
* "ramer" et "cases" est égale à 3.

```python
def dist_hamming(seq1, seq2):
    """
    Calculate the Hamming distance between two sequences.

    Parameters:
    - seq1 (str): The first character string.
    - seq2 (str): The second character string.

    Returns:
    - int: The Hamming distance between the two sequences.
    """
    # Ensure that the sequences are of the same length
    if len(seq1) != len(seq2):
        raise ValueError("Sequences must have the same length.")

    # Initialize the Hamming distance
    hamming_distance = 0

    # Iterate through the characters in the sequences
    for char1, char2 in zip(seq1, seq2):
        # Check if the characters are different
        if char1 != char2:
            hamming_distance += 1

    return hamming_distance

# Test the function with the given sequences
sequence1 = "AGWPSGGASAGLAIL"
sequence2 = "IGWPSAGASAGLWIL"
hamming_dist1 = dist_hamming(sequence1, sequence2)

sequence3 = "ATTCATACGTTACGATT"
sequence4 = "ATACTTACGTAACCATT"
hamming_dist2 = dist_hamming(sequence3, sequence4)

# Print the results
print(f"Hamming distance between '{sequence1}' and '{sequence2}': {hamming_dist1}")
print(f"Hamming distance between '{sequence3}' and '{sequence4}': {hamming_dist2}")
```

#Traducteur chiffre Romains

Ecrire une fonction qui prend un nombre entier compris entre 1 et 3000 inclus, et le convertit en chiffre romains.  
Ci-dessous la correspondance :
https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg
![Exemple de chiffres romains](https://jeretiens.net/wp-content/uploads/2015/12/les_chiffres_romains.jpg)



```python
def int_to_roman(number):
    """
    Convertit un nombre entier en chiffre romain.

    Parameters:
    - number (int): Le nombre entier à convertir (entre 1 et 3000 inclus).

    Returns:
    - str: Le chiffre romain correspondant.
    """
    if not 1 <= number <= 3000:
        raise ValueError("Le nombre doit être compris entre 1 et 3000 inclus.")

    # Définition des symboles romains et de leurs valeurs
    symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]

    roman_numeral = ""

    # Conversion en chiffre romain
    for i in range(len(symbols)):
        while number >= values[i]:
            roman_numeral += symbols[i]
            number -= values[i]

    return roman_numeral

# Test de la fonction avec différents nombres
try:
    input_number = int(input("Entrez un nombre entre 1 et 3000 : "))
    roman_result = int_to_roman(input_number)
    print(f"Le chiffre romain pour {input_number} est : {roman_result}")
except ValueError:
    print("Veuillez entrer un nombre entier valide.")
```
